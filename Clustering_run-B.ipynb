
## RNA

```r
df <- RunUMAP(df, reduction = 'rna_pca', dims = 1:50, assay = 'RNA', reduction.name = 'rna.umap', reduction.key = 'rnaUMAP_')
df <- FindNeighbors(df, reduction = "rna_pca", dims = 1:50, graph.name = "rna_graph.nn")
df <- FindClusters(df, graph.name = "rna_graph.nn", algorithm = 3, resolution = 2, verbose = FALSE)
rnaclusters <- df$seurat_clusters
write.csv(rnaclusters, paste(name, "_RNA_leiden_clusters_run",run,".csv",sep=""))
p3 <- DimPlot(df, reduction = 'rna.umap', label = TRUE, repel = TRUE, label.size = 2.5) + NoLegend()

p3
```

```r
#modularity(df@graphs$RNA_knn, df$seurat_clusters)
g1 <- graph_from_adjacency_matrix(df@graphs$rna_graph.nn , mode = "undirected")
modularity(g1, df$seurat_clusters)
```
